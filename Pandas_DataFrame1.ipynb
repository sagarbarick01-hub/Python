## Q9. Descriptive Statistics

Generate summary statistics using `describe()`. Identify mean sales value, maximum quantity sold, and minimum sales value.

```python
df.describe()
```

```python
mean_sales = df["Sales"].mean()
max_quantity = df["Quantity"].max()
min_sales = df["Sales"].min()

print("Mean Sales:", round(mean_sales, 2))
print("Maximum Quantity Sold:", max_quantity)
print("Minimum Sales:", min_sales)
```

## Q10. Aggregation Analysis

Using the Sales column, calculate total sales, average sales, maximum sales, and minimum sales.

```python
total_sales = df["Sales"].sum()
average_sales = df["Sales"].mean()
maximum_sales = df["Sales"].max()
minimum_sales = df["Sales"].min()

print("Total Sales:", total_sales)
print("Average Sales:", round(average_sales, 2))
print("Maximum Sales:", maximum_sales)
print("Minimum Sales:", minimum_sales)
```
