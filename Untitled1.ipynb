```python
import pandas as pd
import numpy as np

# Q1. Load dataset and display first 5 rows + info
df = pd.read_csv("/content/retail_dataset .csv")  # update path if needed

print("First 5 rows:")
print(df.head())

print("\nDataset Info:")
print(df.info())


# Q2. Missing values
print("\nMissing values per column:")
print(df.isnull().sum())

print("\nTotal missing values:")
print(df.isnull().sum().sum())


# Q3. Replace missing values in Quantity & Price with mean
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].mean())
df['Price'] = df['Price'].fillna(df['Price'].mean())


# Q4. Remove rows where Product Category or Region is missing
df.dropna(subset=['Product Category', 'Region'], inplace=True)


# Q5. Create Revenue column
df['Revenue'] = df['Quantity'] * df['Price']

print("\nDataset with Revenue column:")
print(df.head())


# Q6. Total revenue
total_revenue = np.sum(df['Revenue'])
print("\nTotal Revenue:", total_revenue)


# Q7. Revenue by Product Category
category_revenue = df.groupby('Product Category')['Revenue'].sum()
print("\nRevenue by Category:")
print(category_revenue)


# Q8. Top 3 and Bottom 3 categories
top_3 = category_revenue.sort_values(ascending=False).head(3)
bottom_3 = category_revenue.sort_values().head(3)

print("\nTop 3 Categories:")
print(top_3)

print("\nBottom 3 Categories:")
print(bottom_3)


# Q9. Revenue by Region
region_revenue = df.groupby('Region')['Revenue'].sum()

print("\nRevenue by Region:")
print(region_revenue)

print("\nHighest Revenue Region:", region_revenue.idxmax())
print("Lowest Revenue Region:", region_revenue.idxmin())


# Q10. Monthly revenue + stats
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
df['Month'] = df['Date'].dt.month

monthly_revenue = df.groupby('Month')['Revenue'].sum()
print("\nMonthly Revenue:")
print(monthly_revenue)

# NumPy stats
mean_rev = np.mean(df['Revenue'])
median_rev = np.median(df['Revenue'])
std_rev = np.std(df['Revenue'])

print("\nRevenue Statistics:")
print("Mean:", mean_rev)
print("Median:", median_rev)
print("Standard Deviation:", std_rev)
```
